```python
train = pd.read_csv ('../input/train.csv', )
test = pd.read_csv ('../input/test.csv')
```

```python
#Deleting outliers
train = train.drop(train[(train['GrLivArea']>4000) & (train['SalePrice']<300000)].index)

train_columns_of_interest = train[['OverallQual', 'YearBuilt','1stFlrSF','2ndFlrSF','BedroomAbvGr','GarageArea','GrLivArea','SalePrice']].copy()
test_columns_of_interest = test[['OverallQual', 'YearBuilt','1stFlrSF','2ndFlrSF','BedroomAbvGr','GarageArea','GrLivArea']].copy()
```

```python
test_columns_of_interest['GarageArea'] = test_columns_of_interest['GarageArea'].fillna(0) 
```

```python
#test_columns_of_interest['GarageArea'] = test_columns_of_interest['GarageArea'].fillna(0) 

from sklearn.tree import DecisionTreeRegressor
X_train = train_columns_of_interest.drop(["SalePrice"], axis = 1)
y_train = train_columns_of_interest.SalePrice
X_test = test_columns_of_interest
print(X_train.shape, y_train.shape, X_test.shape)
#
#my_model = DecisionTreeRegressor()
#my_model.fit(X_train, y_train)
#y_pred = my_model.predict(X_test)
```

```python
print(X_train.isnull().sum())
print(X_test.isnull().sum())
```

```python

##from sklearn.preprocessing import Imputer

from sklearn.impute import SimpleImputer
my_imputer = SimpleImputer()
from xgboost import XGBRegressor
X_train=my_imputer.fit_transform(X_train)
X_test=my_imputer.transform(X_test)
```

```python
my_model = XGBRegressor()
my_model.fit(X_train, y_train, verbose=False )
y_pred = my_model.predict(X_test)
```
